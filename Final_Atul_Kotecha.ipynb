```python
#define your own Full Time Employee class here
#Full Time Employee class inherits from Employee class
#the order of the four required positional arguments should be: name,address,vehicle object,salary
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here
import sys
class FullTimeEmployee(Employee):

    def __init__(self, name, add,v, sal):
        if sal <= 0:
            print('Invalid value for salary')
            sys.exit(0)
            
        Employee.__init__(self,name, add,v)
        self.__salary = sal

    def get_salary(self):
        return self.__salary

    def set_salary(self, sal):
        self.__salary = sal
        
    def __str__(self):
        ftp = 'Details of this Full Time Employee Are:\n'
        ftp += super().__str__()
        ftp += "\nAnnual Salary : ${}\n".format(self.__salary)
        return ftp 
        
    def compute_compensation(self):                 # Tax and weekly compensation calculation

        if self.__salary <= 45000 :
            comp = round((self.__salary - (self.__salary*0.18)) /52,2)          
        elif self.__salary <= 82000 :
            comp = round(((self.__salary - (45000*0.18) - ((self.__salary - 45000)*0.28))/52),2)
        else :
            comp = round(((self.__salary - (45000*0.18) - (37000*0.28) - ((self.__salary - 82000)*0.33))/52),2)
        return comp
    
    def compute_reimbursement (self,exp):
        
        if exp <= 10000:
            reim = exp
        else:
            reim = 10000 + 0.5*(exp-10000)
        
        wreim = reim /52
        return wreim
    
```

```python
#define your own Hourly Employee class here
#Hourly Employee class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,hourly rate
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

import sys
class HourlyEmployee(Employee):
    def __init__(self, name, add,v, hrs, rate):
        if (rate <= 0 or hrs< 0 ):
            print('Invalid value for hourly rate and hours worked')
            sys.exit(0)
            
        super().__init__(name, add,v)
        self.__hours_wrkd = hrs
        self.__hourly_rate = rate

    def get_hours_worked(self):
        return self.__hours_wrkd

    def get_hourly_rate(self):
        return self.__hourly_rate            

    def set_hours_worked(self, hw):
        self.__hours_wrkd = hw

    def set_hourly_rate(self, hr):
        self.__hourly_rate = hr

       
    def __str__(self):
        hp = 'Details of this Hourly Employee Are:\n'
        hp += super().__str__()
        hp += "\nHours Worked : {} ; Hourly Rate : ${}\n".format(self.__hours_wrkd, self.__hourly_rate)
        return hp 
        
    def compute_compensation(self):
        if self.__hours_wrkd <= 40 :
            comp = self.__hours_wrkd * self.__hourly_rate
        else :
            comp = ((self.__hours_wrkd- 40)*(1.8*self.__hourly_rate)) + (40* self.__hourly_rate)
        return comp
    
    def compute_reimbursement (self,exp):
        
        if exp <=100:
            reim = exp
        else: 
            reim = 100
            
        return reim   
    
```

```python
#define your own Consultant class here
#Consultant class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

import sys
class Consultant(Employee):
    def __init__(self, name, add,v, hrs, ptype):
        if ( ptype not in (1,2,3) or hrs< 0 ):
            print('Invalid value for project type and hours worked')
            sys.exit(0)
            
        super().__init__(name, add,v)
        self.__hours_wrkd = hrs
        self.__project_type = ptype

    def get_hours_worked(self):
        return self.__hours_wrkd        

    def get_project_type(self):
        return self.__project_type

    def set_hours_worked(self, hw):
        self.__hours_wrkd = hw

    def set_project_type(self, pt):
        self.__project_type = pt

       
    def __str__(self):
        hp = 'Details of this Consultant Are:\n'
        hp += super().__str__()
        hp += "\nHours Worked : {} ; Project Type : {}\n".format(self.__hours_wrkd, self.__project_type)
        return hp 
        
    def compute_compensation(self):
        if self.__project_type ==1 :
            comp = self.__hours_wrkd * 55
        elif self.__project_type == 2 :
            comp = self.__hours_wrkd * 70
        else :
            comp = self.__hours_wrkd * 85
        return comp
    
    
    def compute_reimbursement (self,exp):
        
        if self.__project_type == 1:
            reim = exp
        elif self.__project_type == 2: 
            reim = exp*0.9
        else:
            reim = exp*0.8
            
        return reim 
```

```python
#define your own Management class here
#Management class inherits from both Full Time Employee class and Consultant class
#the order of the six required positional arguments should be: name, address,vehicle object, salary, hours worked, project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class Management(FullTimeEmployee,Consultant):
    def __init__(self, name, add,v, sal,hrs,ptype):         
        FullTimeEmployee.__init__(self,name, add,v,sal)
        Consultant.__init__(self,name,add,v,hrs,ptype)
        
    def __str__(self):  
        mn = 'Details of this Manager Are:\n'
        mn += Employee.__str__(self)
        mn += "\nAnnual Salary : {} Hours Worked : {} ; Project Type : {}\n".format(FullTimeEmployee.get_salary(self),Consultant.get_hours_worked(self),Consultant.get_project_type(self))
        return mn 
    
    
        
    def compute_compensation(self):
        comp = FullTimeEmployee.compute_compensation(self) + Consultant.compute_compensation(self)  #Call compensation method of parent classes
        #print(comp)
        return comp
    
    
    def compute_reimbursement (self,fexp,cexp):
        reim = FullTimeEmployee.compute_reimbursement(self,fexp) + Consultant.compute_reimbursement(self,cexp)
            
        return reim 
```

```python
#complete this get_vehicle_input() function to prepare for the main menu application
#this function is used to ask for vehicle information and then return them: make,model,year,mileage
#this function doesn't take arguement(s)
#no need to handle exceptions of make input and model input
#handle exceptions to make sure year is a 4-digit positive number between 1900 and 2020
#handle exceptions to make sure mileage is a positive number
class year_error(Exception):
    pass
class mil_error(Exception):
    pass

def get_vehicle_input():
    veh_make = input('Enter the vehicle make: ')
    veh_model = input('Enter the vehicle model: ')
    

        
    while True:
        try:
            veh_year = int(input('Enter the vehicle manufactured year:'))
            if (veh_year<1900 or veh_year>2020):
                raise year_error               #Raising error if year not in range
            else : break
        except year_error:
            print('\nPlease enter value between 1900 and 2020')
            continue
                
        except ValueError:                    #Raising Error if input is not integer
            print('\nPlease enter integer')
            continue
             
    while True:
        try:
            veh_mileage = int(input('Enter the vehicle mileage:'))
            if (veh_mileage<0):
                raise mil_error             #Raising error if mileage less than 0
            else : break
        except mil_error:
            print('\nPlease enter value greater than 0')
            continue
                
        except ValueError:
            print('\nPlease enter integer')
            continue
             

    return (veh_make, veh_model, veh_year, veh_mileage)
```

```python
#complete this read_file_data() function to prepare for the main menu application
#this function is used to read information from an existing databse/binary file, which is empdata.dat in the final exam
#this function doesn't take argument(s)
#this function returns a list containing all employee objects stored in the database; each employ object is an element in the list; the list will also be used and updated later
#need to handle exceptions if file doesn't exist
import pickle
def read_file_data():
    my_list = []
    try:
        with (open("empdata.dat", "rb")) as f:
            while True:
                try:
                    my_list.append(pickle.load(f))
                except EOFError:
                    break

        return my_list
    
    except IOError:
        print("File Doesn't exist. Please check the path'")
            
            

        
```

```python
#complete option 5 related code here
#the function is used to show some basic statistics. They are: 
#(1) the number of employees stored in the database
#(2) the highest weekly compensation
#(3) the mean weekly compensation
#(4) the number of employees who have a vehicle with over 100,000 mileage
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#no need to handle exceptions

def run_option5(my_lst):
    
    comp_lst = []
    name = []
    for i in my_lst:
        #print(i)
        #print (i.compensation())
        comp_lst.append(i.compute_compensation())
                   
    for j in range (0,len(my_lst)):
        if my_lst[j].get_vehicle().get_mileage() >= 100000:      #Check for mileage greater than 100000
            name.append(my_lst[j].get_emp_name())                 #Add employee to list if mileage greater than 100000
    #print(name)
        
    print("There are {} employees stored in the database.".format(len(my_lst)))
    print("The highest weekly compensation is ${0:0.2f}.".format(max(comp_lst)))
    print ("The mean weekly compensation of all employee is ${0:0.2f} ". format (sum(comp_lst)/len(comp_lst)))
    print ("There are {} employees having vehicle mileage greater than 100,000. ". format (len(name)))

    
    
```

```python
#complete option 6 related code here
#the function is used to compute weekly reimbursement (assuming the user has the requirerd information)
#the first step is to ask for the name of the employee (you can call run_option4() here)
#the second step is to select one employee from the matched employees
#the third step is to ask for required information 
#this function doesn't return anything
#no need to handle exceptions
def run_option6(emp_list):
    matched_list=run_option4(emp_list)
    choice=int(input("\nWhich employee do you want to check (enter a number)?"))
    emp=matched_list[choice-1]
    etype = emp.__class__.__name__            # Checking type of Employee matching criteria
    #print(etype)
 
    if etype == 'FullTimeEmployee':
        exp = int (input("Enter the Annual Expense for this Employee: "))
        re = emp.compute_reimbursement(exp)
    elif etype == 'HourlyEmployee':
        exp = int (input("Enter the Weekly Expense for this Employee: "))
        re = emp.compute_reimbursement(exp)
    elif etype == 'Consultant':
        exp = int (input("Enter the Weekly Expense for this Employee: "))
        re = emp.compute_reimbursement(exp)
    else:
        fexp = int (input("Enter the Annual Expense for this Employee: "))
        cexp = int (input("Enter the Weekly Expense for this Employee: "))
        re = emp.compute_reimbursement(fexp,cexp)
                
    print('Weekly reimbursment for {0} is ${1:0.2f}'.format(emp.get_emp_name(),re))
```
